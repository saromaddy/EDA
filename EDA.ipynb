```python
# customer data is having 10 null values in state code. 
# dropping rows which contain null values is not an option since the customer is having impact in other csv files like stores.csv. 
# So to keep the sanity of the data, replacing the state code null value with unknown found a better option.
# replacing with mode is also not an option, the state name is Napoli but mode consists of someother state so it creates inconsistency in the data.
```

```python
df["State Code"].fillna('Unknown', inplace=True)
```

```python
df.isnull().sum() #df.customers doesn't have any null values
```

```python
df.to_csv("processed_csv/p_customers.csv")
```
